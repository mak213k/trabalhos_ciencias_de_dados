```python
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

timestamp_seconds = int(time.time())
np.random.seed(timestamp_seconds)


def gerar_dados(qtd=500):
    datas = pd.date_range(start='1/2/2002', periods=qtd, freq='ME')
    receitas = np.random.normal(loc=5000, scale=1500, size=qtd).round(2)
    despesas = np.random.normal(loc=3000, scale=1000, size=qtd).round(2)
    previsto = np.random.normal(loc=4000, scale=1000, size=qtd).round(2)
    real = np.random.normal(loc=3500, scale=1200, size=qtd).round(2)
    df = pd.DataFrame({
        'data': datas,
        'receitas': receitas,
        'despesas': despesas,
        'previsto': previsto,
        'real': real
    })
    return df


def salvar_excel(df, nome='dados_financeiros.xlsx'):
    df.to_excel(nome, index=False)
    print(f"Arquivo salvo: {nome}")


def ler_excel(nome='dados_financeiros.xlsx'):
    df = pd.read_excel(nome)
    print(f"Arquivo carregado: {nome}")
    return df

def exibir_estatisticas(df):
    print("\n=== Cabeçalho ===")
    print(df.head())
    print("\n=== Informações ===")
    print(df.info())
    print("\n=== Estatísticas Descritivas ===")
    print(df.describe())


def plotar_grafico(qtd=500):
    datas = pd.date_range(start='1/2/2022', periods=qtd)
    receitas = np.random.normal(5000, 1500, qtd).cumsum()
    despesas = np.random.normal(4000, 1000, qtd).cumsum()
    df_plot = pd.DataFrame({'Data': datas, 'receitas': receitas, 'despesas': despesas})

    plt.style.use('seaborn-v0_8-darkgrid')
    fig, ax = plt.subplots(figsize=(12, 6))
    ax.plot(df_plot['Data'], df_plot['receitas'], label='Receitas', linewidth=2)
    ax.plot(df_plot['Data'], df_plot['despesas'], label='Despesas', linewidth=2)

    ax.set_title('Análise de Receitas e Despesas ao Longo do Tempo')
    ax.set_xlabel('Data')
    ax.set_ylabel('Valor (R$)')
    ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('R$ {x:,.0f}'))
    plt.legend()
    plt.tight_layout()
    plt.show()


def calcular_mediana(valores):
    valores_ordenados = sorted(valores)
    n = len(valores_ordenados)
    if n % 2 == 0:
        return (valores_ordenados[n//2] + valores_ordenados[n//2 - 1]) / 2
    else:
        return valores_ordenados[n//2]


def calcular_moda(valores):
    contagem = {}
    for valor in valores:
        if valor in contagem:
            contagem[valor] += 1
        else:
            contagem[valor] = 1
    mais_repetido = max(contagem, key=contagem.get)
    return mais_repetido


def analise_numerica(df):
    mediana = calcular_mediana(df['receitas'])
    media = df['receitas'].mean()
    moda = calcular_moda(df['receitas'])
    desvio = df['despesas'].std()

    print(f"\nMediana das Receitas: R$ {mediana:,.2f}")
    print(f"Média das Receitas: R$ {media:,.2f}")
    print(f"Moda das Receitas: R$ {moda:,.2f}")
    print(f"Desvio-Padrão das Despesas: R$ {desvio:,.2f}")


def executar_tudo():
    df = gerar_dados()
    salvar_excel(df, 'dados_financeiros.xlsx')
    df = ler_excel('dados_financeiros.xlsx')
    exibir_estatisticas(df)
   
    analise_numerica(df)
    salvar_excel(df, 'Dados_analisados.xlsx')


def menu():
    while True:
        print("\n=== MENU ===")
        print("1 - Gerar novos dados")
        print("2 - Ver gráfico de receitas e despesas")
        print("3 - Sair do programa")

        try:
            opcao = int(input("Digite o número da opção desejada: "))
           
                          
            if opcao ==1:
                executar_tudo()
                
            elif opcao ==2:
                plotar_grafico()
            
            elif opcao ==3:
                print("Programa encerrado.")
                break
            else:
                print("Opção inválida. Por favor, digite 1, 2 ou 3.")

        except ValueError:
            print("Entrada inválida. Por favor, digite apenas números (1, 2 ou 3).")


menu()
```
